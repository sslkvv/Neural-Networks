```python
import pandas as pd
import json
```

```python
with open('Films_People.txt') as f:
    data = f.read()
js = json.loads(data)
```

```python
dic0 = pd.DataFrame(js)
dic0
```

```python
Name = input("Кому рекомендуем фильм?: ")
```

```python
dictionary = dic0.to_csv('dict_out.csv')  # переводим в csv
dictionary = pd.read_csv('dict_out.csv')
dictionary.drop(columns=dictionary.columns[0], axis= 1 , inplace= True )
dictionary = dictionary.fillna(0)
print(dictionary)
```

```python
from math import sqrt # прописываем ф-ю
def euclid(x, columns, dict):
  d = [1/(1+(sqrt(sum((dictionary[x] - dictionary[i])**2)))) for i in columns]
  values = {}
  for i in range(0, len(columns)):
    if columns[i] != x:
      values[x, columns[i]] = d[i]
  return values
```

```python
dic = {**euclid(Name, dictionary.columns.tolist(), dictionary)}
dic = pd.DataFrame(sorted(dic.items(), key=lambda x: -x[1]))
dic
```

```python
list(dic[0][0])
names = list(dic[0][0])[::-1]
names
```

```python
df = dic0[names]
df = df.fillna(0)
df
```

```python
df = df.drop(df[df[names[0]] == 0.0].index)
df
```

```python
df = df[df[Name] == 0]
df
```

```python
df_sorted = df.sort_values(names[0], ascending=False)
top_3_movies = df_sorted.head(3)
print(top_3_movies.index)
```

```python
print("Рекомендованы к просмотру пользователю", Name,':', ", ".join(top_3_movies.index))
```
