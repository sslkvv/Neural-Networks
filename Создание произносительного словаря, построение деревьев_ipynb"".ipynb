```python
import os

# указываем путь до папки
folder_path = "/content/cta_seg"

# получаем список файлов в папке
files = os.listdir(folder_path)

# выводим список файлов
files = sorted(files)
print(files)
```

```python
sounds = []
words = []
d_new = {}
for file in files:
    # проверяем, что файл имеет нужное расширение
    if file.endswith(".seg_B1"):
        # открываем файл для чтения
        with open(os.path.join(folder_path, file), "r", encoding="utf-8-sig") as B1:
            # читаем содержимое файла
            lines = B1.readlines()
            sounds = [i.strip().split(',') for i in lines[7:]]
    if file.endswith(".seg_Y1"):
        # открываем файл для чтения
        with open(os.path.join(folder_path, file), "r", encoding="cp1251") as Y1:
            # читаем содержимое файла
            lines1 = Y1.readlines()
            words = [[i.strip().split(',')[2],int(i.strip().split(',')[0])] for i in lines1[7:]]
        for i in range(len(words)-1):
          d_new[words[i][0]] = [line[2] for line in sounds if words[i][1]<=int(line[0])<words[i+1][1]]



d_new
```

```python
Unvoiced = {'t':'d','p':'b','k':'g','f':'v','s':'z','zh':'sh'}
for i in d_new.values():
  if i[-1] in Unvoiced.keys():
    i[-1] = Unvoiced[i[-1]]
  if "'" in i[-1]:
    if i[-1][:-1] in Unvoiced.keys():
      #print(i[-1],i[-1][:-1])
      i[-1] = Unvoiced[i[-1][:-1]] +"'"

print(d_new)
```

```python
with open ("data.txt", "w") as fp:
  for x in d_new:
    t = " "
    for i in d_new[x]:
      t = t + " " + i
    fp.write(x + t + "\n")
```
