```python
#считываем sbl файл
def read_sbl(file_name):
    samplerate = 22050
    sampwidth = 2

    with open('/content/sample_data/cta/'+file_name, "rb") as f:
        raw_signal = f.read()
        num_samples = len(raw_signal) // sampwidth
        fmt = str(num_samples) + ("h" if sampwidth == 2 else "c")
        signal = struct.unpack(fmt, raw_signal)
    return signal
```

```python
def mfccs(sbl_files, allophones, start, end, n=13):
    mfccs = []
    all_sbl = []

    for i, sbl_file in enumerate(sbl_files):
        if i == 0 or sbl_file != current_sbl_file:
          signal = read_sbl(sbl_file)
          current_sbl_file = sbl_file

        allophone = allophones[i]
        start_time, end_time = start[i], end[i]
        start_index, end_index = int(float(start_time) / 2), int(float(end_time) / 2)
        sound = np.array(signal[start_index:end_index], dtype=np.float32)
        mfcc = librosa.feature.mfcc(y=sound, sr=22050, n_fft=512, n_mfcc=n)
        mfccs.append((allophone, mfcc))

    return mfccs
```

```python
def mfccs(sbl_files, filename, allophones, start, end, n=13):
    mfccs = []
    all_sbl = []

    for name in filename:
      for sbl_file in sbl_files:
        if name in sbl_file:
            all_sbl.append(sbl_file)
            break

    for i, sbl_file in enumerate(all_sbl):
        if i == 0 or sbl_file != current_sbl_file:
          signal = read_sbl(sbl_file)
          current_sbl_file = sbl_file

        allophone = allophones[i]
        start_time, end_time = start[i], end[i]
        start_index, end_index = int(float(start_time) / 2), int(float(end_time) / 2)
        sound = np.array(signal[start_index:end_index], dtype=np.float32)
        mfcc = librosa.feature.mfcc(y=sound, sr=22050, n_fft=512, n_mfcc=n)
        mfccs.append((allophone, mfcc))

    return mfccs
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state = 200)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#случайный лес
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=20,
                            n_estimators=50,
                            min_samples_split=4,
                            min_samples_leaf=3,
                            max_features=7,
                            random_state=1)
rf.fit(X_train, y_train.ravel())
#rf.fit(X_train, y_train)
print(f'Accuracy составляет {round(rf.score(X_test, y_test)*100,2)}%')
```

```python
#логрегрессия
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(max_iter=500).fit(X_train, y_train.ravel())
print(f'Accuracy составляет {round(LR.score(X_test, y_test)*100,2)}%')
```

```python
rf = RandomForestClassifier()
n_estimators = [10, 50, 100, 200]

# Кросс-валидация для каждого значения количества деревьев
cv_scores = []
for n in n_estimators:
    rf.set_params(n_estimators=n)
    scores = cross_val_score(rf, X_train, y_train, cv=4, scoring='accuracy')
    cv_scores.append(np.mean(scores))

# Построение графика зависимости точности от количества деревьев
import matplotlib.pyplot as plt
plt.plot(n_estimators, cv_scores)
plt.xlabel('Number of trees')
plt.ylabel('Accuracy')
plt.show()

# Выбор оптимального количества деревьев
optimal_n = n_estimators[np.argmax(cv_scores)]
print('Optimal number of trees:', optimal_n)
```
